```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('income.csv')
df.head()
```

```python
plt.scatter(df.Age, df['Income($)'])
plt.xlabel('Age')
plt.ylabel('Income($)')
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age', 'Income($)']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Age, df1['Income($)'], color ='green')
plt.scatter(df2.Age, df2['Income($)'], color ='red')
plt.scatter(df3.Age, df3['Income($)'], color ='black')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color='purple',marker ='*', label ='centroid')
plt.xlabel('Age')
plt.ylabel('Income ($)')
plt.legend()
```

```python
#Preprcessing using min max scaler
scaler = MinMaxScaler()
scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])
scaler.fit(df[['Age']])
df['Age']= scaler.transform(df[['Age']])
df
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age', 'Income($)']])
y_predicted
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Age, df1['Income($)'], color ='green')
plt.scatter(df2.Age, df2['Income($)'], color ='red')
plt.scatter(df3.Age, df3['Income($)'], color ='black')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color='purple',marker ='*', label ='centroid')
plt.xlabel('Age')
plt.ylabel('Income ($)')
plt.legend()
```

# Image segementation using K-means

```python
#read the image
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('image.png')
plt.imshow(image)
plt.show()
```

```python
#convert to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

```python
#reshape the image to a 2D array of pixel and 3 color values (RGB)
pixel_values = image.reshape((-1,3))
#convert to float
pixel_value = np.float32(pixel_values)
print(pixel_value.shape)
```

```python
#define stopping criteria
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
```

```python
# number of cluster (K)
k = 7
_, labels, (centers)= cv2.kmeans(pixel_value, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
```

```python
# convert back to 8-bit values
centers = np.uint8(centers)
#flatten the labels array
labels = labels.flatten()
```

```python
#convert all pixels to the color of the centroids
segmented_image = centers[labels.flatten()]
```

```python
#reshape back to the original image dimensions
segmented_image = segmented_image.reshape(image.shape)
#show the image
plt.imshow(segmented_image)
plt.show()
```
