# Python Pandas Data Structure

### The Pandas provides two data structures for processing the data i.e., Series and DataFrame.

# Series

### It is defined as a one-dimensional array that is capable of storing various data types.The row labels of series are called the index.

# Creating Series from Array:

```python
import pandas as pd  
import numpy as np  
info = np.array(['P','a','n','d','a','s'])  
a = pd.Series(info)  
print(a)
```

# Python Pandas Data Frame

### It is widely used data structure of pandas and works with a two-dimensional array with labeled axes(rows and columns).

### DataFrame is defined as a standard way to store data and has two different indexes i.e., row index and column index. It consists of the following properties:

#### The columns can be heterogeneous types like int,bool, and so on.

#### It can be seen as a dictionary of Series structure where both the rows and columns are indexed. It is denoted as "columns" in case of columns and "index" in case of rows.

## We can create a DataFrame using following ways:

<ul>
    <li>dict</li>
    <li>Lists</li>
    <li>Numpy ndarrays</li>
    <li>Series</li>
</ul>

```python
df = pd.DataFrame()  
print (df)
```

## Create a DataFrame using list


```python
# a list of strings  
x = ['RAM', 'SHYAM']  
  
# Calling DataFrame constructor on list  
df = pd.DataFrame(x)  
print(df)
```

## Create a DataFrame from Dict of ndarrays/ Lists


```python
info = {'ID' :[101, 102, 103],'Department' :['B.Sc','B.Tech','M.Tech',]}  
df = pd.DataFrame(info)  
print (df)
```

## Create a DataFrame from Dict of Series:


```python
info = {'one' : pd.Series([1, 2, 3, 4, 5, 6], index=['a', 'b', 'c', 'd', 'e', 'f']),  
        'two' : pd.Series([1, 2, 3, 4, 5, 6, 7, 8], index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h'])}  
  
d1 = pd.DataFrame(info)  
print (d1)
```

# Column Selection
### We can select any column from the DataFrame. Here is the code that demonstrates how to select a column from the DataFrame.

```python
import pandas as pd  
  
info = {'one' : pd.Series([1, 2, 3, 4, 5, 6], index=['a', 'b', 'c', 'd', 'e', 'f']),  
   'two' : pd.Series([1, 2, 3, 4, 5, 6, 7, 8], index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h'])}  
  
d1 = pd.DataFrame(info)  
print (d1 ['two'])  
print(d1)
```

# Column Addition
### We can also add any new column to an existing DataFrame.

```python
info = {'one' : pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e']),  
   'two' : pd.Series([1, 2, 3, 4, 5, 6], index=['a', 'b', 'c', 'd', 'e', 'f'])}  
  
df = pd.DataFrame(info)  
print(df)
# Add a new column to an existing DataFrame object   
  
print ("Add new column by passing series")  
df['three']=pd.Series([20,40,60],index=['a','b','c'])  
print (df)  
  
print ("Add new column using existing DataFrame columns")  
df['four']=df['one']+df['three']  
  
print (df)
```

# Column Deletion:
### We can also delete any column from the existing DataFrame. 

```python
info = {'one' : pd.Series([1, 2], index= ['a', 'b']),   
   'two' : pd.Series([1, 2, 3], index=['a', 'b', 'c'])}  
     
df = pd.DataFrame(info)  
print ("The DataFrame:")  
print (df)  
  
# using del function  
print ("Delete the first column:")  
del df['one']  
print (df)  
# using pop function  
print ("Delete the another column:")  
df.pop('two')  
print (df)
```

# Selection by Label:
#### We can select any row by passing the row label to a loc function.

```python
info = {'one' : pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e']),   
   'two' : pd.Series([1, 2, 3, 4, 5, 6], index=['a', 'b', 'c', 'd', 'e', 'f'])}  
df = pd.DataFrame(info)  
print(df)
print (df.loc['b'])
```

# Selection by integer location:
### The rows can also be selected by passing the integer location to an iloc function.

```python
info = {'one' : pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e']),  
   'two' : pd.Series([1, 2, 3, 4, 5, 6], index=['a', 'b', 'c', 'd', 'e', 'f'])}  
df = pd.DataFrame(info)  
print (df.iloc[3])
```

# Slice Rows
### It is another method to select multiple rows using ':' operator.

```python
info = {'one' : pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e']),   
   'two' : pd.Series([1, 2, 3, 4, 5, 6], index=['a', 'b', 'c', 'd', 'e', 'f'])}  
df = pd.DataFrame(info)  
print (df[2:5])
```

# Addition of rows:
### We can easily add new rows to the DataFrame using append function. It adds the new rows at the end.

```python
d = pd.DataFrame([[7, 8], [9, 10]], columns = ['x','y'])
print(d)
d2 = pd.DataFrame([[11, 12], [13, 14]], columns = ['x','y'])  
d = d.append(d2)  
print (d)
```

# Deletion of rows:
### We can delete or drop any rows from a DataFrame using the index label. 
### If in case, the label is duplicate then multiple rows will be deleted.

```python
a_info = pd.DataFrame([[4, 5], [6, 7]], columns = ['x','y'])  
b_info = pd.DataFrame([[8, 9], [10, 11]], columns = ['x','y'])  
a_info = a_info.append(b_info)  
print(a_info)
# Drop rows with label 0  
a_info = a_info.drop(0)  
print(a_info)
```
